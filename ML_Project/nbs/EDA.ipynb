```python
categorical_cols = [col for col in df.columns if df[col].dtype == 'object']
numerical_cols = [col for col in df.columns if df[col].dtype != 'object']

print("Categorical Columns:", categorical_cols)
print("Numerical Columns:", numerical_cols)
```

##### Check Unique Values in every Column #####

```python
# Check Unique values in categorical columns
for col in categorical_cols:
    print(f"Unique values in '{col}': {df[col].unique()}")
```

```python
# Vizualize histogram of scores based on gender in subplots

fig, axes = plt.subplots(1, 3, figsize=(15, 5), sharey=True)
sns.histplot(data=df, x='math_score', hue='gender', ax=axes[0], kde=True)
sns.histplot(data=df, x='reading_score', hue='gender', ax=axes[1], kde=True)
sns.histplot(data=df, x='writing_score', hue='gender', ax=axes[2], kde=True)
axes[0].set_title('Math Score Distribution by Gender')
axes[1].set_title('Reading Score Distribution by Gender')
axes[2].set_title('Writing Score Distribution by Gender')
plt.show()
```
